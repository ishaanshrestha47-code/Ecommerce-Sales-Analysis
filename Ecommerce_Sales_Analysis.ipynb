```python
top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)
top_products
```

```python
top_products.plot(kind="bar")
plt.title("top 5 products")
plt.xlabel("products")
plt.ylabel("value")
plt.show()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["month"] = df["Order Date"].dt.month
```

```python
monthy_Sales = df.groupby("month")['Sales'].sum()
months = ["jan","feb","mar","apr","may","june","july","aug","sep","oct","nov","dec"]
monthy_Sales.plot(kind = "bar")
plt.title("monthly wise sales")
plt.xticks(range(12),months,rotation = 20)
plt.show()
```

```python
City_Sales  = df.groupby("City")["Sales"].sum()
City_Sales.plot(kind = "bar")
plt.title("City Wise Sales")
plt.show()
```

```python
City_Sales  = df.groupby("City")["Sales"].sum().sort_values(ascending = False)
City_Sales.head(10).plot(kind = "bar" , figsize = (10,5))
plt.title("top 10 cities by sales")
plt.xlabel("City")
plt.ylabel("Sales")
plt.xticks(rotation=15)
plt.show()
```

```python
df["hour"] = df["Order Date"].dt.hour
hour_Sales = df.groupby("hour")["Sales"].sum()
hour_Sales.plot()
plt.xlabel("hour")
plt.ylabel("Sales")
plt.title("sales by hour")
plt.show()
```

```python
profit = df.groupby("Category")["Profit"].sum()
profit = profit.plot(kind = "bar")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation =15)
plt.show()
```
